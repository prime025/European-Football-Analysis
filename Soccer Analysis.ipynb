```python
import pandas as pd
```

Loading International Soccer results from csv file into a data frame

```python
df = pd.read_csv('results.csv')
df.head()
```

Filtering data to include results after the 2014 World Cup and before the 2018 World Cup

```python
before_results = df[(df['date'] >= '2014-07-14') & (df['date'] <= '2018-06-12')]
before_results.head()
```

Filtering data to only include the 32 teams that played in the 2018 World Cup

```python
wc_teams = ["Argentina", "Australia", "Belgium", "Brazil", "Colombia", "Costa Rica", "Croatia", "Denmark", "Egypt", "England", "France", "Germany", "Iceland", "Iran", "Japan", "Mexico", "Morocco", "Nigeria", "Panama", "Peru", "Poland", "Portugal", "Russia", "Saudi Arabia", "Senegal", "Serbia", "South Korea", "Spain", "Sweden", "Switzerland", "Tunisia", "Uruguay"]
wc_team_results = before_results[(before_results['home_team'].isin(wc_teams)) | (before_results['away_team'].isin(wc_teams))]
wc_team_results.head()
```
